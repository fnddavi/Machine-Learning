```python
x_b = np.c_[np.ones((5,1)),x] #Adiciona x0 = 1 em cada instancia
print(x_b)
```

```python
#x
#y

Y = np.array([y]).T
eta = 0.01 #numero de aprendizados
n_interacoes = 2000
n = 5 #tamanho conjunto de dados
theta = np.randon.randn(2,1)

for interacao in range(n_interacoes):
    gradients = 2/n * x_b.T.dot(x_b.dot(theta) - Y)
    theta = theta - eta * gradients
print("Theta otimizado: " + str(theta))
```

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

iris = datasets.load_iris()
x = iris['data'][:, 3:]  # largura da petala
y = (iris['target'] == 2).astype(np.uint8)  # 1 se iris virginica, 0 se não

log_reg = LogisticRegression()
log_reg.fit(x, y)
x_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(x_new)
plt.plot(x_new, y_proba[:, 1], "g-", label="Iris Virginica")
plt.plot(x_new, y_proba[:, 0], "b--", label="Não Iris Virginica")
plt.xlabel("Largura da pétala (cm)")
plt.ylabel("Probabilidade")
plt.legend(loc="upper left")
plt.title("Probabilidade de ser Iris Virginica vs. Largura da Pétala")
plt.show()
```
